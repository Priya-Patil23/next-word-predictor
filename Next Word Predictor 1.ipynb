```python
import re 
def clean_text(x):
    x = x.lower()
    x = re.sub(r'https?:\/\/.*[\r\n]*' , '' , x , flags = re.MULTILINE)
    x = re.sub(r'\<a href', ' ', x)
    x = re.sub(r'[_"\-;%()|+&=*%.,!?:#$@\[\]/]', ' ', x)
    x = re.sub(r'\'', ' ', x)
    x = re.sub(r"'s\b","",x)
    x = re.sub("[^a-zA-Z]", " ", x)
    x = ' '.join([word for word in x.split() if len(word) >= 3])
    
    text_contract = []
    lemm_text = []
    for i in x.split(" "):
        if i in contractions:
            text_contract.append(contractions[i])
        else:
            text_contract.append(i)
    for i in text_contract:
        lemm_text.append(wordnet.lemmatize(i))
        
    return " ".join(lemm_text)

df_cleaned_data = df['Sentences'].apply(clean_text)
        
    
    
    
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense , LSTM , Embedding
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(df_cleaned_data)
```

```python
tokenizer.word_index
```

```python
new_data  = []

for i in df_cleaned_data:
    tokenize_sentence = tokenizer.texts_to_sequences([i])[0]
    for i in range(1 , len(tokenize_sentence)):
        new_data.append(tokenize_sentence[0:i+1])
        
new_data
        
    
    
```

```python
maxlength = max([len(i) for i in new_data])
```

```python
input_padded = pad_sequences(new_data , maxlen = maxlength , padding = 'pre')
```

```python
input_padded
```

```python
X = input_padded[ : , : -1]
```

```python
Y = input_padded[ : , -1]
```

```python
X
```

```python
model = Sequential()
model.add(Embedding(len(tokenizer.word_index)+ 1 , 100 , input_length = maxlength - 1))
model.add(LSTM(150))
model.add(Dense(len(tokenizer.word_index)+ 1 , activation = 'softmax'))

model.summary()
```

```python
model.compile(optimizer = "Adam" , loss = 'sparse_categorical_crossentropy' , metrics = ['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping , ModelCheckpoint

def early():
    return EarlyStopping(monitor = 'loss' , patience = 10 , min_delta = 0.001 , restore_best_weights = True)

def point(path):
    return ModelCheckpoint(path , save_best_only=True , save_weights_only= True , monitor = 'loss' , save_freq="epoch")

# Let's build two callbacks
```

```python
model.fit(X,Y,epochs = 100 , use_multiprocessing=True, callbacks = [early() , point("next.ckpt")])
```

```python
model.evaluate(X,Y)
```
